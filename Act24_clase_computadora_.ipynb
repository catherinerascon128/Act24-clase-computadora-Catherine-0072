```python
class Computadora:
    def __init__(self, nombre, procesador, ram, almacenamiento, tarjeta_grafica):
        self.nombre = nombre # Usaremos el nombre para identificar cada objeto
        self.procesador = procesador
        self.ram = ram
        self.almacenamiento = almacenamiento
        self.tarjeta_grafica = tarjeta_grafica

    def encender(self):
        print(f"[{self.nombre}]: El sistema está arrancando... ¡Encendida!")

    def ejecutarPrograma(self):
        print(f"[{self.nombre}]: Cargando software principal usando los {self.ram} de RAM...")

    def conectarInternet(self):
        print(f"[{self.nombre}]: Buscando señal Wi-Fi... ¡Conectada a Internet!")

# --- Inicio del programa interactivo ---

nombre1 = input("Introduce el nombre o marca para la Computadora 1: ")
nombre2 = input("Introduce el nombre o marca para la Computadora 2: ")

computadora1 = Computadora(nombre1, "Intel i9", "32GB", "1TB SSD", "RTX 4080")
computadora2 = Computadora(nombre2, "AMD Ryzen 7", "16GB", "500GB SSD", "Radeon RX 6700")


print(f"\n--- Iniciando procesos para {nombre1} ---")
computadora1.encender()
computadora1.ejecutarPrograma()
computadora1.conectarInternet()

print(f"\n--- Iniciando procesos para {nombre2} ---")
computadora2.encender()
computadora2.ejecutarPrograma()
computadora2.conectarInternet()

print("\n--- Tareas completadas ---")
```
